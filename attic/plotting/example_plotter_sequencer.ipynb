### Take only the histogram of the crystal where the beam is focused on

```python
fitting = False

pltdir = '/home/pioneer/shiftdata/plots/'

if fitting:
    fitresult = rd.get_and_plot_calibration_constants()
    fig, ax, figtxt, axtxt, cconstantdict, peakpositiondict, sigmadict = fitresult
    filenamefits = f"cconstants_{filetag}_fits.pdf"
    if not os.path.isfile(pltdir+filenamefits) or overwriteplots:
        fig.savefig(pltdir+filenamefits)
    else: print(f"Plot {filenamefits} already exists and won't be overwritten")
    filenamecc = f"cconstants_{filetag}.pdf"
    if not os.path.isfile(pltdir+filenamecc) or overwriteplots:
        figtxt.savefig(pltdir+filenamecc)
    else: print(f"Plot {filenamecc} already exists and won't be overwritten")

else: #no fitting
    fig,ax = rd.plot_focused_integrals()

plt.show()
```
